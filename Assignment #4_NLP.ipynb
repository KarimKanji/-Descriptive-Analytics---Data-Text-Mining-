### Imports

```python
import wikipedia
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
```

### Assignment

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Preprocessing
def preprocess_text(text):
    # Tokenization
    words = word_tokenize(text.lower())
    # Remove stop words
    words = [w for w in words if not w in stop_words]
    # Lemmatize words
    words = [lemmatizer.lemmatize(w) for w in words]
    # Join words back into text
    text = ' '.join(words)
    return text

# The function that solves the assignment. The function works by preprocessing the given paragraph and question. It then analyses the given paragraph and question and checks to see
# which sentence in the given paragraph best matches the question based on words mentioned. 
def answer_question(paragraph, question):

    # Preprocess paragraph and question
    processed_paragraph = preprocess_text(paragraph)
    processed_question = preprocess_text(question)

    # Tokenize paragraph into sentences
    sentences = sent_tokenize(processed_paragraph)

    # Loop through each sentence and find the one that contains the most matching words with the question
    best_sentence = None
    max_match_count = 0

    for sentence in sentences:
        # Preprocess sentence
        processed_sentence = preprocess_text(sentence)

        # Count number of matching words between question and sentence
        match_count = len(set(processed_question.split()).intersection(processed_sentence.split()))

        # Update best sentence if this one has more matching words than the previous ones
        if match_count > max_match_count:
            best_sentence = sentence
            max_match_count = match_count
            
    # Return the sentence that contains the most matching words as the answer
    return best_sentence


```
